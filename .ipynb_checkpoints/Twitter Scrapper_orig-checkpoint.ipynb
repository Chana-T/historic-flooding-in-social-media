```python
tweet_list = query_tweets(f' "#flood" OR "#flooding" OR "#{city}"  -filter:tweets near:{city} within:10mi',
                               #ADD TIMESTAMP
                              begindate = start_date,
                              enddate = end_date,
                              poolsize = 10,
                              limit=500)
```

```python
df = pd.DataFrame(t.__dict__ for t in tweet_list)
```
